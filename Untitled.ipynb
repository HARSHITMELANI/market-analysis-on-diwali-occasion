# diwali sales analysis 


```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns 
```

```python
df=pd.read_csv('sdiwali.csv' , encoding='unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python

df.drop(['Status','unnamed1'] , axis= 1, inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount'] = df['Amount'].astype('int64')
```

```python
df.info()
```

```python
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

# EDA

GENDER

```python
df.columns
```

```python
ax = sns.countplot(x ='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_g=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender', y='Amount' , data= sales_g)
```

# AGE 

```python
ax = sns.countplot(x ='Age Group',data=df)
```

```python
ax = sns.countplot(x ='Age Group',data=df , hue='Gender')
```

```python
ax = sns.countplot(x ='Age Group',data=df , hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_a =df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group', y='Amount' , data= sales_a)
```

# state 

```python
sales_s =df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.set(rc={'figure.figsize':(35,10)})
sns.barplot(x='State', y='Orders' , data= sales_s)
```

# occupation 

```python
sns.set(rc={'figure.figsize':(25,10)})
ax = sns.countplot(x ='Occupation',data=df )
for bars in ax.containers:
    ax.bar_label(bars)
```

# product categorcategory 

```python
sns.set(rc={'figure.figsize':(25,10)})
ax = sns.countplot(x ='Product_Category',data=df )
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_p =df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='Product_Category', y='Amount' , data= sales_p)
```
